```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
f = open("./20230531_CVD_3076_map/fits_out/FIT/det_1_0_1.txt_FITDIR/det_1_0_1.txt_2.1.00001.1.fit", "r")

#ignore first three lines - s will serve as garbage bin
s = f.readline()
s = f.readline()
s = f.readline()

#because there is a space between the [ and the first number, lines can be processed using split(" ")
energy = f.readline().split(" ")[3:-1]
ydata = f.readline().split(" ")[3:-1]
yfit = f.readline().split(" ")[3:-1]
bg = f.readline().split(" ")[3:-1]

f.close()

del s
```

```python
for i in range(0, len(energy)):
    energy[i] = float(energy[i])
    ydata[i] = float(ydata[i])
    yfit[i] = float(yfit[i])
    bg[i] = float(bg[i])
    
energy = np.array(energy)
ydata = np.array(ydata)
yfit = np.array(yfit)
bg = np.array(bg)
```

```python
f, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (10, 4))

ax1.plot(energy, bg)
ax1.grid(True)
ax1.axis([min(energy), max(energy), 0, 175])
ax1.set_xlabel("Energy, keV")
ax1.set_ylabel("Intensity")

ax2.semilogy(energy, bg)
ax2.axis([min(energy), max(energy), 3, 175])
ax2.grid(True)
ax2.set_xlabel("Energy, keV")
ax2.set_ylabel("Intensity")

plt.show()
```
