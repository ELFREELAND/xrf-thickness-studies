```python
import numpy as np
from matplotlib import pyplot as plt
import glob
import csv
```

```python
#path_label = "20230531_CVD_3158"

#f1 = open("./" + path_label + "_map/fits_out/FIT/det_1_0_1.txt_FITDIR/det_1_0_1.txt_2.1.00001.1.fit", "r")
f1 = open("./cvd3158 fitting experiments/fits_out_E/FIT/det_1_0_0.txt_FITDIR/det_1_0_0.txt_2.1.00001.1.fit", "r")

#ignore first three lines - s will serve as garbage bin
s = f1.readline()
s = f1.readline()
s = f1.readline()

#because there is a space between the [ and the first number, lines can be processed using split(" ")
energy1 = np.array([ float(i) for i in f1.readline().split(" ")[3:-1] ])
ydata1 = np.array([ float(i) for i in f1.readline().split(" ")[3:-1] ])
yfit1 = np.array([ float(i) for i in f1.readline().split(" ")[3:-1] ])
bg1 = np.array([ float(i) for i in f1.readline().split(" ")[3:-1] ])

f1.close()

del s
```

```python
energy1 = np.array([])
ydata1 = np.array([])
yfit1 = np.array([])
bg1 = np.array([])

with open("./cvd3158 fitting experiments/3158_100_snip_60_voigt.csv", "r") as f:
    rd = csv.reader(f)
    for i in rd:
        if i[0] == "channel":
            pass
        else:
            energy1 = np.append(energy1, float(i[1]))
            ydata1 = np.append(ydata1, float(i[2]))
            yfit1 = np.append(yfit1, float(i[3]))
            bg1 = np.append(bg1, float(i[4]))
```

```python
f, ax = plt.subplots(figsize = (10, 6))

ax.semilogy(energy1, ydata1)
ax.semilogy(energy1, yfit1)
ax.semilogy(energy1, bg1)
ax.grid(True)
ax.axis([min(energy1), max(energy1), 1, 40000])
ax.set_xlabel("Energy, keV")
ax.set_ylabel("Intensity")
ax.set_title("CVD3158 measurement 1-0-0, fit with 60-channel SNIP + voigt peaks")

plt.show()
```
