```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
path_label = "20230531_CVD_3076"

f1 = open("./" + path_label + "_map/fits_out/FIT/det_1_0_1.txt_FITDIR/det_1_0_1.txt_2.1.00001.1.fit", "r")

#ignore first three lines - s will serve as garbage bin
s = f1.readline()
s = f1.readline()
s = f1.readline()

#because there is a space between the [ and the first number, lines can be processed using split(" ")
energy1 = [ float(i) for i in f1.readline().split(" ")[3:-1] ]
ydata1 = [ float(i) for i in f1.readline().split(" ")[3:-1] ]
yfit1 = [ float(i) for i in f1.readline().split(" ")[3:-1] ]
bg1 = [ float(i) for i in f1.readline().split(" ")[3:-1] ]

f1.close()

del s
```

```python
path_label = "20230531_CVD_3076"

f2 = open("./" + path_label + "_map/fits_out/FIT/det_1_0_0.txt_FITDIR/det_1_0_0.txt_2.1.00001.1.fit", "r")

#ignore first three lines - s will serve as garbage bin
s = f2.readline()
s = f2.readline()
s = f2.readline()

#because there is a space between the [ and the first number, lines can be processed using split(" ")
energy2 = [ float(i) for i in f2.readline().split(" ")[3:-1] ]
ydata2 = [ float(i) for i in f2.readline().split(" ")[3:-1] ]
yfit2 = [ float(i) for i in f2.readline().split(" ")[3:-1] ]
bg2 = [ float(i) for i in f2.readline().split(" ")[3:-1] ]

f2.close()

del s
```

```python
energy1 = np.array(energy1)
ydata1 = np.array(ydata1)
yfit1 = np.array(yfit1)
bg1 = np.array(bg1)

energy2 = np.array(energy2)
ydata2 = np.array(ydata2)
yfit2 = np.array(yfit2)
bg2 = np.array(bg2)
```

```python
f, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (10, 4))

ax1.plot(energy1, bg1)
ax1.plot(energy2, bg2)
ax1.grid(True)
ax1.axis([min(energy1), max(energy1), 0, 175])
ax1.set_xlabel("Energy, keV")
ax1.set_ylabel("Intensity")

ax2.semilogy(energy1, bg1)
ax2.semilogy(energy2, bg2)
ax2.axis([min(energy2), max(energy2), 3, 175])
ax2.grid(True)
ax2.set_xlabel("Energy, keV")
ax2.set_ylabel("Intensity")

plt.show()
```
